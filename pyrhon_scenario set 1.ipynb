```python
#5,A text-processing tool reverses a given word or sentence for formatting purposes.
# Write logic to take a word or sentence as input and produce its reversed version.
def reverse_user_string():
    # 1. Get input from the user
    user_input = input("Please enter a word or sentence to reverse: ")
    
    # 2. Reverse the string using slice notation [::-1]
    # The slice [::-1] creates a new string that is the reverse of the original.
    reversed_string = user_input[::-1]
    
    # 3. Print the original and reversed string
    print("\n--- Result ---")
    print(f"Original String: {user_input}")
    print(f"Reversed String: {reversed_string}")
    print("--------------")

reverse_user_string()
```

```python
#10 A program filters out only even numbers from a given list.
# Write logic to extract and return only the even numbers from a list.
def get_even_numbers():
    # Taking input from user as space-separated numbers
    num_list = list(map(int, input("Enter numbers separated by space: ").split()))
    
    even_list = []
    
    for i in num_list:
        if i % 2 == 0:
            even_list.append(i)
    
    return even_list

get_even_numbers()
```
