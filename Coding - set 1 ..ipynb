```python
#4.Write a Python function to find the factorial of 5. Expected Output:120
def fact():
    num=int(input("Enter the number:"))
    a=1
    for i in range(1,num+1):
        a=a*i
#        p=print("Factorial of", num, "is", a)
    return a

fact()
```
