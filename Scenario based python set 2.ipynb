```python
"""
2.
A digital locker requires users to enter a numerical passcode. As part of a security feature, the system checks the sum of the digits of the passcode.
 Write logic to compute the sum of the digits of a given number.
"""
def sum_of_digits():
    num = input("Enter a number: ")
    total = 0
    for digit in num:
        total += int(digit)
    print("Sum of digits:", total)

sum_of_digits()
```

```python
"""
3.
A mobile payment app uses a simple checksum validation where reversing a transaction ID helps detect fraud.
 Write logic to take a number and return its reverse.
"""
num = input("Enter a number: ")   # Read the input number as a string
rev = num[::-1]                  # Reverse the string using slicing
reversed_num = int(rev)          # Convert the reversed string back to a number

print("Reversed number:", reversed_num)
```

```python
"""
4.
In a secure login system, certain features are enabled only for users with prime-numbered user IDs.
 Write logic to check if a given number is prime.
"""
num = int(input("Enter a number: "))

# Step 1: If number < 2 → Not Prime
if num < 2:
    print("Not Prime")
else:
    is_prime = True

    # Step 2: Loop from 2 to square root of number
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:        # Found a divisor
            print("Not Prime")
            is_prime = False
            break

    # Step 3: If no divisors found → Prime
    if is_prime:
        print("Prime")
```

```python
"""
5.
A scientist is working on permutations and needs to calculate the factorial of numbers frequently.
 Write logic to find the factorial of a given number using recursion.
"""
num = int(input("Enter a number: "))

fact = 1

for i in range(1, num + 1):
    fact = fact * i   # multiply each number

print("Factorial of", num, "is", fact)
```

```python
"""
6 A unique lottery system assigns ticket numbers where only Armstrong numbers win the jackpot.
 Write logic to check whether a given number is an Armstrong number.
"""

num = int(input("Enter a number: "))

# Step 1: Count number of digits
digits = len(str(num))

# Step 2: Initialize sum
total = 0

# Step 3: Make a copy of the number
temp = num

# Step 4: Process each digit
while temp > 0:
    digit = temp % 10
#    print("digit:",digit)
    total += digit ** digits   # Raise digit to power and add
#    print(total)
    temp //= 10                # Remove last digit
#    print(temp)

# Step 5: Compare
if total == num:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")

```

```python
"""
7
A password manager needs to strengthen weak passwords by swapping the first and last characters of user-generated passwords.
 Write logic to perform this operation on a given string.
"""
s = input("Enter a string: ")

if len(s) < 2:
    print("Modified string:", s)
else:
    modified = s[-1] + s[1:-1] + s[0]
    print("Modified string:", modified)
```

```python
"""
8
A low-level networking application requires decimal numbers to be converted into binary format before transmission.
 Write logic to convert a given decimal number into its binary equivalent
"""
# Step 1: Read the input decimal number
num = int(input("Enter a decimal number: "))

# Step 2: Initialize an empty string for binary
binary = ""

# Step 3: Handle the case when number is 0
if num == 0:
    binary = "0"
else:
    n = num
    while n > 0:
        remainder = n % 2               # Divide by 2 and get remainder
        binary += str(remainder)        # Add remainder to binary string
        n = n // 2                      # Update number by dividing by 2

    # Step 4: Reverse the binary string
    binary = binary[::-1]

# Step 5: Print the binary representation
print("Binary equivalent:", binary)
```

```python
"""
9.
A text-processing tool helps summarize articles by identifying the most significant words.
 Write logic to find the longest word in a sentence.
"""
# Step 1: Read the input sentence
sentence = input("Enter a sentence: ")

# Step 2: Split the sentence into individual words
words = sentence.split()

# Step 3: Initialize a variable to store the longest word
longest_word = ""

# Step 4: Loop through each word
for word in words:
    # Step 5: Update longest_word if current word is longer
    if len(word) > len(longest_word):
        longest_word = word

# Step 6: Print the longest word
print("The longest word is:", longest_word)

```

```python
"""
10.
A plagiarism detection tool compares words from different documents and checks if they are anagrams (same characters but different order).
 Write logic to check whether two given strings are anagrams.
"""
# Step 1: Read the two input strings
str1 = input("Enter the first string: ")
str2 = input("Enter the second string: ")

# Step 2: Remove spaces and convert to lowercase
str1 = str1.replace(" ", "").lower()
str2 = str2.replace(" ", "").lower()

# Step 3: Sort the characters of both strings
sorted_str1 = sorted(str1)
sorted_str2 = sorted(str2)

# Step 4: Compare sorted strings
if sorted_str1 == sorted_str2:
    print("Anagram")
else:
    print("Not an Anagram")
```
